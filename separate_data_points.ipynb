```python
# extract the organic carbon burial at each point

orgbur = np.zeros(3)
calbur = np.zeros(3)

for pts in range(3):
    out = Lines[244*pts:244*(pts+1)]
    SolBal = out[69:75]
    
    # writing to file
    file = open('SolBal.txt', 'w')
    file.writelines(SolBal)
    file.close()
    
    # read data to pandas
    SolBaldf = pd.read_csv('SolBal.txt', delim_whitespace=True)
    
    # add burial of fast and slow organic matter
    orgbur[pts]=SolBaldf.FOrg[3]-SolBaldf.SOrg[3]
    calbur[pts]=SolBaldf.Cal[3]
```
